Translating reviews from french to english

```python
from translate import Translator
translator = Translator(from_lang = 'fr', to_lang = 'en')
translation = translator.translate('Salut')
print(translation)
```

```python
train['avis_en'] = train['avis'].apply(lambda x: translator.translate(str(x)))
train
```

Counting adverbes and ajectives in reviews.

```python
text = 'I am a test and this is a word.'
tokens = nltk.word_tokenize(text)
tokens
```

```python
tags = nltk.pos_tag(tokens)
tags
```
